```python
import pandas as pd
df = pd.read_csv("imdb_reviews.csv")
corpus = df['review'].tolist()
corpus[:5]
```

```python
import re
import spacy

nlp = spacy.load("en_core_web_sm")


def clean_text(documents: list[str]):
    cleaned_docs = []
    for doc in documents:
        doc_text = re.sub(r"[^\w\s]", "", doc.lower())
        doc_nlp = nlp(doc_text)
        filtered_text = [token.text for token in doc_nlp if not token.is_stop and token.text.strip()]
        cleaned_docs.append(filtered_text)

    return cleaned_docs

cleaned_corpus = clean_text(corpus[:1000])
cleaned_corpus[:5]
```

```python
from gensim.models import Word2Vec

model = Word2Vec(
    sentences=cleaned_corpus,
    vector_size=32,
    min_count=2,
    window=10,
    workers=4,
    compute_loss=True,
    epochs=100,
    alpha=0.001
)

model
```

```python
model.wv.most_similar("incredible")
```
